This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  12


Find the method performs best:


```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Assuming X and y are your features and labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create models
models = {
    "KNN": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(max_iter=1000, solver='liblinear'),  # Increased max_iter and changed solver
    "Decision Tree": DecisionTreeClassifier(),
    "SVM": SVC()
}

# Dictionary to store results
results = {}

# Evaluate models using cross-validation
for name, model in models.items():
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=10)  # Using 10-fold cross-validation
    results[name] = cv_scores.mean()

# Compare the results
best_model = max(results, key=results.get)
best_score = results[best_model]

# Visualize cross-validation results (Bar Chart)
plt.figure(figsize=(10, 6))
plt.bar(results.keys(), results.values(), color='skyblue')
plt.title('Cross-Validation Accuracy Scores for Different Models', fontsize=16)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.show()

print("Cross-validation results (accuracy):")
for name, score in results.items():
    print(f"{name}: {score:.4f}")

print(f"\nBest model: {best_model} with an accuracy of {best_score:.4f}")

# Train the best model and evaluate on the test set
best_model_instance = models[best_model]
best_model_instance.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = best_model_instance.predict(X_test_scaled)

# Calculate accuracy and other metrics
test_accuracy = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy of {best_model}: {test_accuracy:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title('Confusion Matrix for Best Model', fontsize=16)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.show()

# Classification Report as a heatmap
report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()

# Visualize classification report as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(report_df.iloc[:-1, :].T, annot=True, cmap='Blues', fmt='.2f', cbar=True)
plt.title(f'Classification Report for {best_model}', fontsize=16)
plt.ylabel('Metrics', fontsize=12)
plt.xlabel('Classes', fontsize=12)
plt.show()
```
